```python
import json

data = {}
with open('Esox_lucius.json') as file:
    data = json.load(file)
```

```python
data['dna'].keys()
```

```python
len(data['cds'])
```

```python
total_length = {}

dna_types = data.keys()

for dna_type in dna_types:
    total_length[dna_type] = 0
    print(total_length)
    for value in data[dna_type].values():
        total_length[dna_type] += value['all']
```

```python
total_length
```

```python
a_values = {}

for key, values in data['cds'].items():
    a_values[key] = values['A']
```

```python
len(a_values)
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(20,10))

#plt.plot(list(a_values.values()))
plt.scatter(range(len(a_values)),list(a_values.values()), alpha=0.1)
```

```python
gc_values = {}
total_lengths = {}

for key, values in data['cds'].items():
    gc_values[key] = values['GC']
    total_lengths[key] = values['all']
    
figure1 = plt.figure(figsize=(20,10))

plt.scatter(list(total_lengths.values()),list(gc_values.values()), alpha=0.1)
```

```python
at_values = {}
total_lengths = {}

for key, values in data['cds'].items():
    at_values[key] = values['A'] + values['T']
    total_lengths[key] = values['all']
    
figure2 = plt.figure(figsize=(20,10))

plt.scatter(list(total_lengths.values()),list(at_values.values()), alpha=0.1)
```

```python
gc_values = {}
at_values = {}
total_lengths = {}

for key, values in data['cds'].items():
    gc_values[key] = values['GC']
    at_values[key] = values['A'] + values['T']
    total_lengths[key] = values['all']
    
figure3 = plt.figure(figsize=(20,10))

plt.scatter(list(total_lengths.values()),list(gc_values.values()), alpha=0.1, color='y')
plt.scatter(list(total_lengths.values()),list(at_values.values()), alpha=0.1, color='r')
```
