```python
from ftplib import FTP
import gzip

ftp = FTP('ftp.ensembl.org')
ftp.login()
```

```python
ftp.cwd('/pub/current_fasta/gorilla_gorilla/cds')
```

```python
with open('file.fa.gz', 'wb') as file:
    ftp.retrbinary('RETR Gorilla_gorilla.gorGor4.cds.all.fa.gz', file.write)
```

```python
handle = gzip.open('file.fa.gz')
with open('gorilla_gorilla.fa', 'wb') as out:
    for line in handle: 
        out.write(line)
```

```python
from Bio.SeqIO.FastaIO import FastaIterator

record_ids = []
records = []

with open("gorilla_gorilla.fa") as handle:
     for record in FastaIterator(handle):
            record_ids.append(record.id)
            records.append(record)
```

```python
str(records[0].description)
```

```python
data = {}

for record in records:
    data[record.id] = {'A': 0, 'G': 0, 'C': 0, 'T': 0, 'N': 0}
    for letter in str(record.seq):
        data[record.id][letter] += 1
```

```python
import csv

with open('data.csv', 'w') as csvfile:
    writer = csv.writer(csvfile, delimiter = "\t")
    writer.writerow(['ID', 'A', 'G', 'C', 'T', 'N'])
    for key, values in data.items():
        writer.writerow([key, values['A'], values['G'], values['C'], values['T'], values['N']])
```
